```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler()
scaler.fit(df_market_data)
new_scaled = scaler.transform(df_market_data)
new_scaled
```

```python

ncr_df = pd.DataFrame(new_scaled, columns=['price_change_percentage_24h', 'price_change_percentage_7d', 'price_change_percentage_14d', 'price_change_percentage_30d', 'price_change_percentage_60d', 'price_change_percentage_200d',	'price_change_percentage_1y'])

nc = list(df_market_data.index)

ncr_df['coin_id'] = nc
ncr_df.set_index('coin_id', inplace=True)

ncr_df.head(10)
```

---

### Find the Best Value for k Using the Original Data.

```python

list_of_kv = list(range(1,12))
list_of_kv
```

```python

iv_set = []
for k in list_of_kv:
#
    km_m = KMeans(n_clusters=k, random_state=40)
    
    
    km_m.fit(new_scaled)
    
   
    iv_set.append(km_m.inertia_)

print(iv_set)
```

```python
ski = {}
for k in list_of_kv:
    km_m = KMeans(n_clusters=k, random_state=40)
    km_m.fit(new_scaled)
    ski[k] = km_m.inertia_

elb_df = pd.DataFrame(list(ski.items()), columns=['k-value', 'inertia'])
elb_df
```

```python

elb_df.hvplot.line(
    x="k-value", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k,
    line_color='darkmagenta',
    line_dash='dotted',
    line_width=9 
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** According to the elbow curve, it seems the best value for k is 4. 

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python

new_k_mod = KMeans(n_clusters=4,random_state=40)
```

```python

new_k_mod.fit(ncr_df)
```

```python

predict_four = new_k_mod.predict(new_scaled)


predict_four
```

```python

chosen_cluster = ncr_df.copy()
```

```python

chosen_cluster['predicted_cluster'] = predict_four


chosen_cluster.head(10)
```

```python

chosen_cluster.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="predicted_cluster", 
    hover_cols = 'coin_id', 
    marker = 'd', 
    size = 300,
    color=['teal', 'darkorchild', 'navy', 'red']
)
```

---

### Optimize Clusters with Principal Component Analysis.

```python

pcamod = PCA(n_components=3)
```

```python

cpca = pcamod.fit_transform(chosen_cluster)

cpca
```

```python

ret_var = pcamod.explained_variance_ratio_
ret_var
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total of the explained variance of the three principal compenets is 0.89188968. 

```python

new_pca_df = pd.DataFrame(cpca, columns = ['PC1', 'PC2', 'PC3'])

new_pca_df

new_pca_df['coin_id']= list(chosen_cluster.index)


new_pca_df = new_pca_df.set_index('coin_id')


new_pca_df.head(10)
```

---

### Find the Best Value for k Using the PCA Data

```python

new_k_valv = list(range(1,12))
new_k_valv
```

```python

second_init = []


for i in new_k_valv:
    k_m_two = KMeans(n_clusters=i, random_state=1)
    k_m_two.fit(new_pca_df)
    second_init.append(k_m_two.inertia_)
```

```python

elb_two = {"k":new_k_valv, "inertia":second_init}

elb_two_df = pd.DataFrame(elb_two)
elb_two_df
```

```python

elb_two_df.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k,
    line_color='palevioletred',
    line_dash='dashed',
    line_width=9 
)
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** According to this it seems like its 4. 


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No its the same.
